```python
# Linear Regression with Gradient Descent

# Importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (12.0, 9.0)

# Preprocessing Input data
from google.colab import files
uploaded = files.upload()
data = pd.read_excel('data.xlsx')
 # Replace 'data.csv' with the path to your dataset
X = data.iloc[:, 0]  # Assuming the first column contains the independent variable
Y = data.iloc[:, 1]  # Assuming the second column contains the dependent variable

# Plotting the data
plt.scatter(X, Y)
plt.title("Scatter Plot of Data")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

# Building the model
m = 0  # Initial slope
c = 0  # Initial intercept
L = 0.0001  # Learning Rate
epochs = 1000  # Number of iterations to perform gradient descent
n = float(len(X))  # Number of data points

# Performing Gradient Descent
for i in range(epochs):
    Y_pred = m * X + c  # Predicted value of Y
    Dm = (-2 / n) * sum(X * (Y - Y_pred))  # Partial derivative with respect to m
    Dc = (-2 / n) * sum(Y - Y_pred)  # Partial derivative with respect to c
    m = m - L * Dm  # Updating m
    c = c - L * Dc  # Updating c
    if i % 100 == 0:  # Print progress every 100 iterations
        print(f"Epoch {i}: m = {m}, c = {c}")

# Making predictions
Y_pred = m * X + c

# Plotting the results
plt.scatter(X, Y, label="Original Data")
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red', label="Regression Line")  # Regression line
plt.title("Linear Regression using Gradient Descent")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
